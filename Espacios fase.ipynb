$$
\frac{dY}{dt}=AY,\qquad \text{donde }A=\begin{pmatrix}

0 & 2
\end{pmatrix}
$$

### Sumideros

Cuando tenemos eigenvalores negativos tenemos **sumideros**. Consideremos el sistema

$$
\frac{dY}{dt}=CY,\qquad \text{donde }C=\begin{pmatrix}
-1 & 0\\
0 & -4
\end{pmatrix}
$$

```julia
#=su campo de direcciones; 
en este caso α=-3, β=0, γ=0 y δ=2 =#

xs = collect(-2.5:0.1:2.5)
ys = collect(-2.5:0.1:2.5)

X,Y = np.meshgrid(xs,ys)

U = -X
V = -4Y
title(L"Espacio Fase: $y(t)$ vs $x(t)$")
axhline(0,color="gray")
axvline(0,color="gray")
streamplot(xs,ys,U,V,color=U, cmap=plt.cm.autumn)
xlabel(L"$x(t)$")
ylabel(L"$y(t)$")
#savefig("hola")
```

El otro sistema es:

$$

\end{pmatrix}
$$

```julia
#=su campo de direcciones; 
en este caso α=-3, β=0, γ=0 y δ=2 =#

xs = collect(-2.5:0.1:2.5)
ys = collect(-2.5:0.1:2.5)

X,Y = np.meshgrid(xs,ys)

U = -2X - 2Y
V = -X - 3Y
title(L"Espacio Fase: $y(t)$ vs $x(t)$")
#axhline(0,color="gray")
#axvline(0,color="gray")
streamplot(xs,ys,U,V,color=U, cmap=plt.cm.autumn)
xlabel(L"$x(t)$")
ylabel(L"$y(t)$")

sumidero(x) = -1/2*x
plot(xs,[sumidero(x) for x in xs], label="Solución al equilibrio")
legend()
savefig("sumidero2")
```

### Fuentes

Cuando tenemos ambos eigenvalores positivos tenemos una **fuente**. Consideremos el sistema

$$
\frac{dY}{dt}=EY,\qquad \text{donde }E=\begin{pmatrix}
2 & 2\\
1 & 3
\end{pmatrix}
$$

```julia
#=su campo de direcciones; 
en este caso α=-3, β=0, γ=0 y δ=2 =#

xs = collect(-2.5:0.1:2.5)
ys = collect(-2.5:0.1:2.5)

X,Y = np.meshgrid(xs,ys)

U = 2X + 2Y
V = X + 3Y
title(L"Espacio Fase: $y(t)$ vs $x(t)$")
#axhline(0,color="gray")
#axvline(0,color="gray")
streamplot(xs,ys,U,V,color=U, cmap=plt.cm.autumn)
xlabel(L"$x(t)$")
ylabel(L"$y(t)$")

sumidero(x) = -1/2*x
plot(xs,[sumidero(x) for x in xs], label="Solución al equilibrio")
legend()
savefig("fuente")
```

### Espacio fase de modelo de competencia de especies

El sistema es el siguiente
$$
\begin{align*}
\frac{dx}{dt}&=2x\left(1-\frac{x}{2}\right)-xy\\
\frac{dy}{dt}&=3y\left(1-\frac{y}{3}\right)-2xy
\end{align*}
$$

```julia
xs = collect(0:0.1:4)
ys = collect(0:0.1:4)

X,Y = np.meshgrid(xs,ys)

U = 2X - X.*X - X.*Y
V = 3Y - Y.*Y -2X.*Y
title("Competencia de especies")
#axhline(0,color="gray")
#axvline(0,color="gray")
streamplot(xs,ys,U,V,color=U, cmap=plt.cm.autumn)
xlabel(L"$x(t)$")
ylabel(L"$y(t)$")

scatter(1,1, label = L"(1,1)")
scatter(0,3, label = L"(0,3)")
scatter(2,0, label = L"(2,0)")
scatter(0,0, label = L"(0,0)")

legend()

savefig("competencia de epecies")
```
